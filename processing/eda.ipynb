## 9. Quantification des valeurs aberrantes – Méthode IQR

Pour confirmer la présence de valeurs aberrantes observées dans le boxplot, nous utilisons la méthode de l’Intervalle Interquartile (IQR).

Le calcul montre un nombre important de valeurs aberrantes (575 574 observations), ce qui confirme la forte dispersion de la variable Ntop.

Ces valeurs élevées peuvent correspondre :

- à des régions très peuplées,
- à des pathologies fortement répandues,
- ou à des caractéristiques structurelles du dataset.

Il est donc nécessaire de réfléchir à leur traitement : les conserver si elles reflètent une réalité métier, ou appliquer une transformation (comme le logarithme) afin de stabiliser la variable pour la modélisation.

```python
Q1 = df["Ntop"].quantile(0.25)
Q3 = df["Ntop"].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df["Ntop"] < lower) | (df["Ntop"] > upper)]
print("Nombre d'outliers détectés :", len(outliers))
```

## 15.Association entre pathologie (niv.1) et région – V de Cramér

Le calcul du V de Cramér entre `patho_niv1` et `region` donne une valeur proche de 0.

Cela indique qu’il n’existe pas d’association forte entre le type de pathologie (niveau 1) et la région au niveau global.

Autrement dit, la répartition des grandes catégories de pathologies semble relativement homogène entre les régions, sans dépendance catégorielle marquée.

```python


crosstab = pd.crosstab(df["patho_niv1"], df["region"])

chi2 = ss.chi2_contingency(crosstab)[0]
n = crosstab.sum().sum()
phi2 = chi2/n
r,k = crosstab.shape
cramer_v = np.sqrt(phi2/min(k-1, r-1))

cramer_v
```

## 16.Association entre pathologie (niv.1) et région – V de Cramér & p-value

La p-value de 1.0 confirme l'absence totale de dépendance catégorielle entre 
patho_niv1 et region. Cela s'explique par la structure du dataset : toutes les 
pathologies sont présentes dans toutes les régions de manière homogène.

La région sera néanmoins conservée comme feature car les disparités 
de volume de cas (Ntop) entre régions sont significatives.

```python
chi2_stat, p_value, dof, expected = ss.chi2_contingency(crosstab)

print(f"V de Cramér : {round(cramer_v, 4)}")
print(f"p-value : {p_value}")
print("Conclusion :", "Lien significatif" if p_value < 0.05 else "Pas de lien détectable")
```

## 18.Agrégation des données et recalcul de la prévalence régionale

Afin d’adapter l’analyse à la problématique métier, les données sont regroupées par :

- année
- région
- pathologie niveau 1

L’objectif est de recalculer une prévalence correcte à un niveau agrégé.

En effet, la variable prev fournie dans le dataset est calculée à un niveau plus fin (département, âge, sexe).  
Pour comparer correctement les régions, il est nécessaire de recalculer la prévalence selon la formule :

Prevalence = ΣNtop / ΣNpop

Ce recalcul garantit une comparaison cohérente et statistiquement valide entre territoires et périodes.

Cette étape est essentielle car elle aligne l’analyse exploratoire avec la problématique business (évolution régionale des pathologies).

```python
df_grouped = (
    df.groupby(["annee","region","patho_niv1"])
    .agg({"Ntop":"sum","Npop":"sum"})
    .reset_index()
)

df_grouped["prevalence"] = df_grouped["Ntop"] / df_grouped["Npop"]
```

## 19. Comparaison entre la prévalence fournie et la prévalence recalculée

Cette étape compare la prévalence fournie dans le dataset (`prev_original`) avec la prévalence recalculée à partir des données agrégées (`prevalence = Ntop / Npop`).

L’objectif est de vérifier la cohérence des indicateurs et de s’assurer que les valeurs de prévalence correspondent bien au ratio entre le nombre de cas et la population.

On observe que certaines différences peuvent apparaître, ce qui peut s’expliquer par :
- des méthodes de calcul différentes (moyenne vs ratio agrégé),
- des arrondis,
- ou des valeurs manquantes dans certaines observations.

Cette vérification permet de valider la qualité des données avant la phase de modélisation.
La prévalence recalculée est exprimée en proportion (0–1), tandis que la variable prev du dataset semble exprimée en pourcentage (0–100). Les écarts observés proviennent donc en partie de cette différence d’échelle.

```python
prev_original = (
    df.groupby(["annee","region","patho_niv1"])["prev"]
    .mean()
    .reset_index()
)

comparison = df_grouped.merge(
    prev_original,
    on=["annee","region","patho_niv1"],
    how="left"
)

comparison.rename(columns={"prev":"prev_original"}, inplace=True)

comparison.head()
```
